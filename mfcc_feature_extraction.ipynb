```python
df=pd.read_csv("/Users/dhwanibhavankar/Desktop/SML_Project/Data.csv")
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
files_folder = '/Users/dhwanibhavankar/Downloads/ICBHI_final_database'
```

```python
# Create empty lists to store counts and total rows
crackles_counts = []
wheezes_counts = []
total_rows_counts = []
```

```python
# Iterate over the rows in the DataFrame
for _, row in df.iterrows():
    # Form the corresponding text file name
    common_filename = os.path.splitext(row['File Names'])[0]
    text_file = os.path.join(files_folder, f"{common_filename}.txt")

    # Read the text file and count crackles, wheezes, and total rows
    try:
        with open(text_file, 'r') as file:
            lines = file.readlines()
            crackles_count = sum(1 for line in lines if line.split('\t')[2] == '1')  # Assuming crackles are in the third column
            wheezes_count = sum(1 for line in lines if line.split('\t')[3] == '1')   # Assuming wheezes are in the fourth column
            total_rows_count = len(lines)

            # Append counts to the respective lists
            crackles_counts.append(crackles_count)
            wheezes_counts.append(wheezes_count)
            total_rows_counts.append(total_rows_count)
    except FileNotFoundError:
        # Handle the case where the text file is not found
        print(f"Text file not found for audio file: {row['File Names']}")
```

```python
df['Crackles Count'] = crackles_counts
```

```python
df['Wheezes Count'] = wheezes_counts
df['Total Rows Count'] = total_rows_counts
```

```python
df.head()
```

```python
df.describe()
```

```python
# Calculate the percentage of crackles and wheezes
df['Crackles Percentage'] = (df['Crackles Count'] / df['Total Rows Count']) * 100
df['Wheezes Percentage'] = (df['Wheezes Count'] / df['Total Rows Count']) * 100
```

```python
df = df.drop('Wheezes Percentage',axis=1)
```

```python
import os
import pandas as pd
import librosa
```

```python
def extract_mfcc(file_path, n_mfcc=13):
    y, sr = librosa.load(file_path, sr=None)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return mfccs.mean(axis=1)  # Take the mean along the time axis
```

```python
# Folder path
folder_path = '/Users/dhwanibhavankar/Downloads/ICBHI_final_database'
```

```python
# Initialize an empty DataFrame
columns = ['File Name'] + [f'MFCC_{i+1}' for i in range(13)]  # Assuming n_mfcc=13
df_features = pd.DataFrame(columns=columns)
```

```python
# Iterate through files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.wav'):
        file_path = os.path.join(folder_path, file_name)
        mfcc_features = extract_mfcc(file_path)
        df_features = df_features.append({'File Name': file_name, **dict(zip(columns[1:], mfcc_features))}, ignore_index=True)
```

```python
df_features
```

```python
df_features_mfcc = df_features
```
